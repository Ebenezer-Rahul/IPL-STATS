```python
class Team :
    def __init__(self, team_id, points, team_name, num_of_matches, num_Won, num_Lost):
        self.team_id = team_id
        self.points = points
        self.team_name = team_name
        self.num_of_matches = num_of_matches
        self.num_Won = num_Won
        self.num_Lost = num_Lost
        pass
```

```python
class Match:
    def __init__(self, match_id, toss_winner_id, toss_decision, date, city, venue, team1, team2, winner, result, result_margin, umpire1,umpire2, team1_runs, team2_runs, team1_wickets, team2_wickets, man_of_match):
        self.match_id = match_id
        self.toss_winner_id = toss_winner_id
        self.toss_decision = toss_decision
        self.date = date
        self.city = city
        self.venue = venue
        self.team1 = team1
        self.team2 = team2
        self.winner = winner
        self.result = result
        self.result_margin = result_margin
        self.umpire1 = umpire1
        self.umpire2 = umpire2
        self.team1_runs = team1_runs
        self.team2_runs = team2_runs
        self.team1_wickets = team1_wickets
        self.team2_wickets = team2_wickets
        self.man_of_Match = man_of_match
        pass
```

```python
class Player:
    def __init__(self, player_id, team_id, player_name):
        self.player_id = player_id
        self.team_id = team_id
        self.player_name = player_name
        self.runs = 0
        self.wickets = 0
        self.best_score = 0
        self.dismissals = 0
        self.innings_played = 0
        self.number_of_balls_faced = 0
        self.number_of_sixes = 0
        self.number_of_fours = 0
        self.num_of_balls_bowled = 0
        self.number_of_boundries = 0
        self.run_outs = 0
        self.run_outs_taken = 0
        self.num_catches = 0
        self.strike_rate = 0.0
        self.num_runs_given = 0
        pass
```

```python
count = 0
player_exists = {}
players = {}
team_exists = {'nan':'nan'}
teams = {}
match_exists = {}
matches = {}
```

```python
team_id = 1
player_id = 1
for index,data in data_2016.iterrows():
    if(team_exists.get(data['team1']) == None):
        new_team = Team(team_id,0, data['team1'], 0,0,0)
        team_exists[data['team1']] = team_id
        teams[team_id] = new_team
        team_id += 1
        print(new_team.team_name)
        count += 1
        
for index,data in data_2016.iterrows():
    if(player_exists.get(data['batsman']) == None):
        new_player = Player(player_id, team_exists[data['batting_team']], data['batsman'])
        player_exists[data['batsman']] = player_id
        players[player_id] = new_player
        player_id += 1
        
        
    if(player_exists.get(data['non_striker']) == None):
        new_player = Player(player_id, team_exists[data['batting_team']], data['non_striker'])
        player_exists[data['non_striker']] = player_id
        players[player_id] = new_player
        player_id += 1
        
    if(player_exists.get(data['bowler']) == None):
        new_player = Player(player_id, team_exists[data['bowling_team']], data['bowler'])
        player_exists[data['bowler']] = player_id
        players[player_id] = new_player
        player_id += 1
    

print(count)
```

```python
num_pl = 0
for key in players.keys():
    print(f'{players[key].player_name}, {teams[players[key].team_id].team_name}')
    num_pl += 1
print(num_pl)
```

```python
num_mat = 0
c_ = 0
for index,data in data_2016.iterrows():
    if(match_exists.get(data['id']) == None):
        #, data['result'], data['result_margin'], data['umpire1'],data['umpire2'], 0,0,0,0)
        new_match = Match(data['id'], team_exists[data['toss_winner']], data['toss_decision'], data['date'], data['city'], data['venue'], team_exists[data['team1']], team_exists[data['team2']],team_exists[data['winner']], data['result'], data['result_margin'], data['umpire1'],data['umpire2'], 0,0,0,0,data['player_of_match'])
        match_exists[data['id']] = 1
        num_mat += 1
        print(data['winner'])
        print(num_mat)
        matches[data['id']] = new_match
        
    
    if(data['batting_team'] == teams[matches[data['id']].team1].team_name):
        matches[data['id']].team1_runs += int(data['total_runs'])
        
    if(data['batting_team'] == teams[matches[data['id']].team2].team_name):
        matches[data['id']].team2_runs += int(data['total_runs'])
        
    if(data['bowling_team'] == teams[matches[data['id']].team1].team_name):
        matches[data['id']].team2_wickets += int(data['is_wicket'])
        
    if(data['bowling_team'] == teams[matches[data['id']].team2].team_name):
        matches[data['id']].team1_wickets += int(data['is_wicket'])
    
    players[player_exists[data['batsman']]].runs += int(data['batsman_runs'])
    #print(players[player_exists[[data['bowler']]]].num_of_balls_bowled)
    players[player_exists[data['bowler']]].num_of_balls_bowled += 1
    #print(type(data['extras_type']))     
    c_ += 1
    if((data['extras_type'] == 'nan') or type(data['extras_type']) == type(0.1) or data['extras_type'] == 'legbyes'or data['extras_type'] == 'no_ball'):
        players[player_exists[data['batsman']]].number_of_balls_faced += 1
            
            
    if(data['batsman_runs'] == '4' or data['batsman_runs'] == '6'):
        players[player_exists[data['batsman']]].number_of_boundries += 1
    
    if(data['batsman_runs'] == '4'):
        players[player_exists[data['batsman']]].number_of_fours += 1
    
    if(data['batsman_runs'] == '6'):
        players[player_exists[data['batsman']]].number_of_sixes += 1
    
    if(data['player_dismissed'] != 'nan' or type(data['player_dismissed']) == type(0.0)):
        players[player_exists[data['batsman']]].dismissals += 1
   
    print(data['dismissal_kind'])
    if(data['dismissal_kind'] != 'run out' and int(data['is_wicket']) > 0):
        players[player_exists[data['bowler']]].wickets += 1
    
    
    
    if(data['dismissal_kind'] == 'catch'):
        players[player_exists[data['feilder']]].num_catches += 1
    
    if(data['dismissal_kind'] == 'run out'):
        players[player_exists[data['batsman']]].run_outs += 1
        #print(type(data['fielder']))
        #players[player_exists[data['fielder']]].run_outs_taken += 1
    
print(f'c_= {c_}')
```
